# Replace missing values: Numerical

```python
def replaceNullNum(col):
  mean = combined_df[col].mean()
  combined_df[col].fillna(value=mean, inplace=True)
```

```python
for col in num_columns:
  replaceNullNum(col)
```

# Replace missing values: Categorical


```python
def replaceNullCat(col):
    mode = combined_df[col].mode().iloc[0]
    combined_df[col].fillna(value= mode, inplace=True)
```

```python
for col in cat_columns:
  replaceNullCat(col)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split

new_df = merged_df.drop(["Year"], axis = 1)

# Assuming 'merged_df' is your DataFrame
train_data, test_data = train_test_split(new_df, test_size=0.2, random_state=123)
X_train = train_data.drop('STAT_CAUSE_CODE', axis=1)
y_train = train_data['STAT_CAUSE_CODE']

# Separate features and labels for testing data
X_test = test_data.drop('STAT_CAUSE_CODE', axis=1)
y_test = test_data['STAT_CAUSE_CODE']

# Assuming 'continuous_features' and 'discrete_features' are lists of feature column names
continuous_features = ["population", "estimated_houses", "allstate_pif_fraction", "allstate_policies_in_force", "LONGITUDE", "LATITUDE"]
discrete_features = ["county_fips"]

# Define the preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), continuous_features),
        ('cat', 'passthrough', discrete_features)
    ]
)

# Combine preprocessing with the KNN classifier in a pipeline
# Use the DTW distance function as the metric
knn_model = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', KNeighborsClassifier(n_neighbors=15, metric="euclidean"))  
])


# Fit the model on the training data
knn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn_model.predict(X_test)

# Evaluate the model on the test set
test_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy on the test set: %.3f' % test_accuracy)
```
